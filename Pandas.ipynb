```python
print(data.bedrooms) # Accessing a column as an attribute
```

```python
print(data['bedrooms']) # Accessing values using the indexing ([]) operator
```

```python
print(data['bedrooms'][0]) # Retrieving the value of the 'bedrooms' column at index 0
```

```python
# Accessing a subset of data from the DataFrame
print(data['bedrooms'][0:10])
```

```python
print(data['bedrooms'].iloc[0:10])
```

```python
# Selecting the first row of data
print(data.iloc[0])
```

```python
# Selecting a column with iloc
print(data.iloc[:, 3])
```

```python
print(data.iloc[:, 3:5])
```

```python
# Using a list for selection
print(data.iloc[[0, 1, 2], 0])
```

```python
# Negative numbers for selection (last five elements)
print(data.iloc[-3:])
```

```python
# Accessing the value of the 'bedrooms' column at index 0
print(data['bedrooms'][0])

# Using loc for label-based selection
print(data.loc[0, 'bedrooms'])
```

```python
print(data.loc[:, ['bedrooms', 'price', 'floors']])
```

```python
# Accessing a subset of data using conditions
data[data['price'] > 500000]
```

```python
# Using multiple conditions for selection
data[(data['price'] > 500000) & (data['bedrooms'] == 3)]
```

```python
# Assigning values to specific locations
data.loc[data['price'] > 500000, 'high_value'] = 1
```

```python
# Check and print missing values
print(data.isnull())
```

```python
# Print the count of missing values in each column
print(data.isnull().sum())
```

```python
# Fill NaN values in the entire DataFrame with 0
data.fillna(0)
```

```python
# Fill NaN values in the 'bedrooms' column with 0
data['bedrooms'].fillna(0)
```

```python
# Drop rows with any NaN values, returning a new DataFrame
data.dropna(axis=0, inplace=False)
```

```python
# Drop columns with any NaN values, modifying the original DataFrame
data.dropna(axis=1, inplace=True)
data
```

```python
# Assigning a value (0) to the 'floors' column
data['floors'] = 0
data
```

```python
# Assign values to the 'floors' column using a list comprehension
data['floors']=[i for i in range(data.shape[0])]
data
```

```python
# Display concise information about the DataFrame using the info() function
data.info()
```

```python
# Generate statistical summaries for numeric columns in the DataFrame using the describe() function
data.describe()
```

```python
# Print the mean and standard deviation of the 'price' column
print("Mean Price:", data.price.mean())
print("Standard Deviation Price:", data.price.std())

# Print unique values in the 'bedrooms' column
print("Unique Bedrooms:", data.bedrooms.unique())
```
